```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
df = pd.read_csv("cleaned_mental_health_dataset.csv")
df.head()
```

```python
# Check missing values
print(df.isnull().sum())

# Check for duplicates
print("Duplicate rows:", df.duplicated().sum())

# Basic stats
print(df.describe())

# Check categorical columns
print(df['Gender'].value_counts())
print(df['Exercise Level'].value_counts())
print(df['Diet Type'].value_counts())
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Stress Level", order=df["Stress Level"].value_counts().index, palette="pastel")
plt.title("Stress Level Distribution")
plt.xlabel("Stress Level")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Mental Health Condition", order=df["Mental Health Condition"].value_counts().index, palette="muted")
plt.title("Mental Health Condition Distribution")
plt.xlabel("Condition")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Exercise Level", y="Happiness Score", palette="Set2")
plt.title("Happiness Score vs Exercise Level")
plt.show()
```

```python
# Average Happiness by Exercise Level
exercise_happy = df.groupby("Exercise Level")["Happiness Score"].mean()
print(exercise_happy)

# Average Stress by Diet Type
diet_stress = df.groupby("Diet Type")["Stress Level"].value_counts(normalize=True)
print(diet_stress)
```

```python
df['Diet Group'] = df['Diet Type'].replace({'Balanced':'Healthy', 'Vegetarian':'Healthy', 'Vegan':'Healthy',
                                            'Junk Food':'Unhealthy','Keto':'Unhealthy'})
```

```python
# Percentage of low-exercise people with low happiness
low_happy_pct = df[(df['Exercise Level']=='Low') & (df['Happiness Score']<=4)].shape[0] / df.shape[0] * 100
print(f"{low_happy_pct:.1f}% of low-exercise participants report low happiness")
```

```python
corr = df[['Sleep Hours','Screen Time per Day (Hours)','Social Interaction Score','Happiness Score']].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data
exercise_happiness = df.groupby("Exercise Level")["Happiness Score"].mean().reset_index()

# Plot
plt.figure(figsize=(8,5))
sns.barplot(data=exercise_happiness, x="Exercise Level", y="Happiness Score", palette="Set2")
plt.title("Average Happiness Score by Exercise Level", fontsize=14)
plt.xlabel("Exercise Level", fontsize=12)
plt.ylabel("Average Happiness Score", fontsize=12)

# Add value labels on top
for index, row in exercise_happiness.iterrows():
    plt.text(index, row["Happiness Score"]+0.02, round(row["Happiness Score"],2), ha='center', fontsize=11)

plt.show()
```

```python
# Calculate proportion
stress_diet = df.groupby(["Diet Type","Stress Level"]).size().unstack(fill_value=0)
stress_diet = stress_diet.div(stress_diet.sum(axis=1), axis=0)  # convert to proportion

# Plot stacked bar chart
stress_diet.plot(kind="bar", stacked=True, figsize=(10,6), colormap="Pastel1")

plt.title("Stress Level Distribution by Diet Type", fontsize=14)
plt.xlabel("Diet Type", fontsize=12)
plt.ylabel("Proportion", fontsize=12)
plt.legend(title="Stress Level", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Calculate average Happiness for each Exercise Level & Diet Type
combined_avg = df.groupby(["Exercise Level","Diet Type"])["Happiness Score"].mean().reset_index()

# Plot
plt.figure(figsize=(12,6))
sns.barplot(data=combined_avg, x="Diet Type", y="Happiness Score", hue="Exercise Level", palette="Set2")
plt.title("Average Happiness by Exercise Level & Diet Type", fontsize=14)
plt.xlabel("Diet Type", fontsize=12)
plt.ylabel("Average Happiness Score", fontsize=12)
plt.ylim(4.8, 5.7)  # optional: zoom in for better visual difference
plt.legend(title="Exercise Level", bbox_to_anchor=(1.05, 1), loc='upper left')

# Add value labels
for i, row in combined_avg.iterrows():
    plt.text(x=i%5 + (i//5)*0.05,  # rough positioning
             y=row["Happiness Score"] + 0.01,
             s=f"{row['Happiness Score']:.2f}",
             fontsize=9,
             ha='center')

plt.tight_layout()
plt.show()
```
